```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient()
database = client['exercices']
collection = database['arbres']
```

```python
df_arbres = pd.read_csv("./data/arbresremarquablesparis.csv")
df_arbres.head(6)
```

# Nettoyage des données

```python
l_d = df_arbres.to_dict(orient='records')
```

# Import des données

```python
from pymongo import MongoClient
client = MongoClient("mongo:27017")
```

```python
db_ar = client.ar
collection_ar = db_ar['asantanadb']
```

```python
collection_ar.delete_many({})
collection_ar.insert_many(l_d)
```

# Afficher les arbres dans l'arondissement bois de boulogne

```python
query = collection_ar.find({"ARRONDISSEMENT":"BOIS DE BOULOGNE"})
list(query)
```

# Afficher les arbres les plus larges

```python
cur = collection_ar.find({}).sort([("CIRCONFERENCE EN CM", -1)])
list(cur)
```

# Compter le nombre d'arbres pour chaque espèce

```python
req = collection_ar.aggregate([{"$group": {"_id" : "$ESPECE", "Total" : {"$sum" : 1}}}])
list(req)
```

# Afficher les arbres qui mesurent plus de 30 mètres uniquement

```python
req = collection_ar.find({"$and":[{'HAUTEUR EN M':{'$gte':30.0}}]})
req.count()
list(req)
```
